```python
import pandas as pd
import numpy as np
```

**Series**

*#Pandas Series - Like a Table Column - 1d Array*

```python
my_series = [5,9,12,27]
my_var = pd.Series(my_series)
print(my_var)
```

*#Print specific value*

```python
my_var[2]
```

*#Labels using index argument*

```python
my_index = ["A", "B", "C", "D"]
my_var2 = pd.Series(my_series, index = my_index)
print(my_var2)
```

```python
my_var2["B"]
```

*#Key Dictionery*

```python
cars = {"Tesla":12, "Mercedes":42, "BMW":3}
my_var3 = pd.Series(cars)
print(my_var3)
```

```python
my_var3["Tesla"]
```
