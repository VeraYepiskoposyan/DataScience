```python
my_df4 = pd.read_csv('dog_data_short.csv')
my_df4
```

```python
#Add column From list
gender = ["Male", "Female", "Male", "Male", "Female"]
my_df4["Gender"] = gender
my_df4
```

```python
# Add Default Values
my_df4["Alive/Dead"] = [True] * len(my_df4)
my_df4
```

```python
# Add NaN Values
my_df4["Show Dog"] = [np.nan] * len(my_df4)
my_df4
```

```python
# Add column with .insert() - Allows position 
my_df4.insert(1, "Adopted", [True] * len(my_df4), True) #True allows duplicates
my_df4
```

```python
# Add column with assign() - creates new df
my_df5 = my_df4.assign(Horse = [False] * len(my_df4))
my_df5
```
