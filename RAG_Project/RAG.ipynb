```python
documents = [
    "Armenia has a parliamentary political system.",
    "Machine learning models require training data.",
    "Policy decisions impact economic development.",
    "Support Vector Machines are supervised models.",
    "Elections are part of democratic systems."
]
```

```python
from sentence_transformers import SentenceTransformer

model = SentenceTransformer('all-MiniLM-L6-v2')
```

```python
doc_embeddings = model.encode(documents)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np

query = "How do political systems work?"

query_embedding = model.encode([query])

similarities = cosine_similarity(query_embedding, doc_embeddings)

top_index = np.argmax(similarities)

print("Most relevant document:")
print(documents[top_index])
```

```python
top_k = 2
indices = np.argsort(similarities[0])[-top_k:][::-1]

for i in indices:
    print(documents[i])
```
