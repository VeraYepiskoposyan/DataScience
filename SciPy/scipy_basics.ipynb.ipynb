**Special Functions**

```python
import numpy as np
import scipy.special as special
```

```python
#Computes the cube root of each element
print("Cube roots: ", special.cbrt(np.array([64, 125, 729])))
```

```python
#Computes 10 ** x
print("exp10: ", special.exp10([2,3]))
```

```python
#Combinations: special(N, k) where 
#N is the number of objects 
# k is the number of objects being picked (order doesnt matter)
print("Combination: ", special.comb(10,3))
```

```python
# Permutation: special.perm(N, k), where
#N is the number of objects 
# k is the number of objects being picked (order matters)
print("Permutation: ", special.perm(10,3))
```

```python
# logsumexp: Computes the log of the sum of exponentials of input element
# input: [x1, x2, x3x, x4]
# output: loge((e ** x1) + (e ** x2) + (e ** x3) + (e ** x4))
print("logsumexp: ", special.logsumexp([1, 2, 3, 4]))
```

```python
# Lambert w function
# Defined as the inverse of x * (e ^ x)
# W(z) is such that z = W(z) * (e ^ W(z)) where z is a complex number
print("W(1): ", special.lambertw(1))
```

**Interpolation**

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = np.arange(11)
Y = np.array([2.0, 1.9, 1.7, 1.5, 0.5, 0.0, 0.8, 2.0, 0.9, 0.2, 2.0])

plt.plot(X, Y, "o:")
plt.show()
```

```python
from scipy.interpolate import interp1d
# 'Linear', 'Nearest', 'Zero', 'Slinear', 'Quadratic', 'Cubic'
predict = interp1d(X, Y, kind="linear")

X2 = np.linspace(0, 10, 100)
Y2 = np.array([predict(x) for x in X2])

plt.plot(X, Y, "o:")
plt.plot(X2, Y2, "ro:")
plt.show()

print(f"X: 1.5, Y: {predict(1.5)}")
```

```python
import numpy as np
import scipy.linalg as linalg
```

```python
# 3x + 4y = 10
# 2x + 5y = 9

a = np.array([
    [3,4],
    [2,5]
])
b = np.array([10,9])
print("Solution:", linalg.solve(a,b))
```

```python
array = np.array([
    [1,2,3],
    [4,5,6],
    [8,8,9]
])
print("Numpy array:")
print(array)
print()

# finding determinant
determinant = linalg.det(array)
print("Determinant:", determinant)
print()
```

```python
# Finding matrix inverse
inverse_array = linalg.inv(array)
print("Inverse: ")
print(inverse_array)
print()
```

```python
print("Matrix - inverse product: ")
print(array @ inverse_array)
print()
```

```python
# Obtaining Eigenvalues and Eigenvectors
array2 = np.array([
    [1,-2,0],
    [-2,1,0],
    [0,0,3]
])

eg_val, eg_vec = linalg.eig(array2)
print(eg_val)
print(eg_vec)
```

```python
# Singular value decomposition
# array2 = PD(P_inverse)
P = eg_vec
P_inverse = linalg.inv(eg_vec)

D = []
for index, eg_value in enumerate(eg_val):
    row = np.zeros((3))
    row[index] = eg_value
    D.append(row)
D = np.array(D)

print(P @ D @ P_inverse)
```

**Integration**

```python
import math 
from scipy.integrate import quad
```

```python
def f(x: float) -> float:
    return math.sin(x)

area, error = quad(f, 0, math.pi / 2)
print(f"Area = {area}")
print(f"Error = {error}")
```

**Double Integration**

```python
import math 
from scipy.integrate import dblquad
```

```python
def f(x,y):
    return x ** 2 + y ** 2

area, error = dblquad(f, -5, 5, -10, 10)
print(f"Area = {area}")
print(f"Error = {error}")
```

```python
import math 
from scipy.integrate import nquad
```

```python
def f(x, y, z):
    return x * y * z

result, error = nquad(f, [[0, 2], [-2, 0], [1, 2]])
print(f"Result = {result}")
print(f"Error = {error}")
```

**Optimization**

```python
import numpy as np
from scipy.optimize import minimize
```

```python
def objective(x):
    a = x[0]
    b = x[1]
    return (a - 4)**2 + (b - 3)**2 + 10

x0 = np.array([200,500])
solution = minimize(objective, x0, method ="SLSQP")
print(solution)
```

**Equality Constraint**

```python
def objective(x):
    a = x[0]
    b = x[1]
    return a**2 + b**2

def constraint1(x):
    a = x[0]
    b = x[1]
    return a + b - 100

cons = [
    {
        "type": "eq",
        "fun": constraint1
    }
]

x0 = np.array([200,500])
solution = minimize(objective, x0, method ="SLSQP", constraints = cons)
print(solution)
```

```python
def objective(x):
    a = x[0]
    b = x[1]
    return a**2 + b**2

#def constraint1(x):
    #a = x[0]
    #b = x[1]
    #return a + b - 100

def constraint2(x):
    a = x[0]
    b = x[1]
    return a * b - 100

cons = [
    #{
        #"type": "eq",
        #"fun": constraint1
    #},
        {
        "type": "ineq",
        "fun": constraint2
    }
]

x0 = np.array([200,500])
solution = minimize(objective, x0, method ="SLSQP", constraints = cons)
print(solution)
```

**Bounds**

```python
def objective(x):
    a = x[0]
    b = x[1]
    return a**2 + b**2

bound = np.array([10,50])
bounds = (bound, bound)

x0 = np.array([200,500])
solution = minimize(objective, x0, method ="SLSQP", bounds = bounds)
print(solution)
```
