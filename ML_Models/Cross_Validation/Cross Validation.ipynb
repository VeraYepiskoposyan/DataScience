```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import cross_val_score, StratifiedKFold, train_test_split, KFold
```

```python
dataset = pd.read_csv('adult.csv')
dataset.isnull().sum()
```

```python
dataset.head()
```

```python
dataset['marital-status'].unique()
```

```python
#Converting Sex value into 0 or 1
dataset["gender"] = dataset["gender"].map({"Male" : 0, "Female" : 1})

dataset["income"] = dataset["income"].map({
    "<=50K": 0,
    ">50K": 1
})

#Create Married Column - Binary Yes(1) or No(0)
# dataset["marital-status"] = dataset["marital-status"].replace(["Never-married", "Divorced", "Separated", "Widowed"], 0)
# dataset["marital-status"] = dataset["marital-status"].replace(["Married-civ-spouse", "Married-spouse-absent", "Married-AF-spouse"],1).infer_objects(copy=False)

# dataset["marital-status"] = dataset["marital-status"].astype(int)

dataset["marital-status"] = dataset["marital-status"].isin(
    ["Married-civ-spouse", "Married-spouse-absent", "Married-AF-spouse"]
).astype(int)
```

```python
dataset.drop(labels = ["workclass", "education", "occupation", "relationship", "race", "native-country"], axis = 1, inplace = True)
dataset.head()
```

```python
y = dataset["income"]
dataset.drop('income', axis = 1)

X_train, X_test, y_train, y_test = train_test_split(dataset, y, test_size = 0.2, random_state = 42)
```

```python
models = []
models.append(("LR", LogisticRegression()))
models.append(('KNN', KNeighborsClassifier()))
```

```python
results = dict()
for name, model in models:
    kfold = KFold(n_splits=10)
    cv_results = cross_val_score(model, X_train, y_train, cv = kfold, scoring = "accuracy")
    results[name] = (cv_results.mean(), cv_results.std())

print()
print("name     results.mean     results.std")
for key, value in results.items():
    print(key, value)
```
